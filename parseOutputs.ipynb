```python
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python

def extract_values_from_file(filename, search_string, unit):
    with open(filename, 'r') as file:
        lines = file.readlines()

    pattern = re.compile(fr'{search_string}\s+=\s+([-\d.E+]+) {unit}')
    numerical_values = []

    for line in lines:
        match = pattern.search(line)
        if match:
            value_str = match.group(1)
            value_float = float(value_str)
            numerical_values.append(value_float)

    return numerical_values
```

```python
error = 'Energy error'
units = 'Ry'
values = np.array([])
for i in range(3):
    values = np.append(values, extract_values_from_file('output_'+str(i+1), error, units)[1:])
plt.title('PBE0 vc_relax geomtery 0 '+error)
plt.semilogy(values)
plt.ylabel(error + ' (' +units+')')
plt.xlabel('Iteration')
plt.axhline(y=1e-4, color='r', linestyle='--', linewidth=2, label='y = 0.5')
```

```python
error = 'Gradient error'
units = 'Ry/Bohr'
values_0 = np.array([])
for i in range(3):
    values_0 = np.append(values_0, extract_values_from_file('output_'+str(i+1), error, units)[1:])
plt.title('PBE0 vc_relax geomtery 0 '+error)
plt.semilogy(values_0)
plt.ylabel(error + ' (' +units+')')
plt.xlabel('Iteration')
plt.axhline(y=1e-3, color='r', linestyle='--', linewidth=2, label='y = 0.5')
```
